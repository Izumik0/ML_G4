```python
import pandas as pd
import numpy as np
from natsort import natsorted
```

```python
data = pd.read_csv('Database/lifetimes_data_simple.csv')
data['name_dir'] = data['seq'].astype(str)+'_'+data['top'].astype(str)+'_'+data['lay'].astype(str)+'_'+data['copy'].astype(str)
print(data.head())
```

Ogarnięcie chi kątów

```python
data=pd.read_csv('dihedrals_medoids_1c_20n/dihedrals_medoids_new_chi.csv')
meta_cols=[c for c in data.columns if not c.startswith('D')]
angels_cols=[c for c in data.columns if c.startswith('D')]
angel_data=data[angels_cols]
DG_cols=[c for c in angel_data.columns if c.startswith('DG')]
DT_cols=[c for c in angel_data.columns if c.startswith('DT')]
DG_data=angel_data[DG_cols].rename(columns=lambda x: x.replace('DG', 'K_') if x.startswith('DG') else x).copy()
DT_data=angel_data[DT_cols].rename(columns=lambda x: x.replace('DT', 'K_') if x.startswith('DT') else x).copy()
new_ang_data=DG_data.combine_first(DT_data)
ang_cols=[c for c in new_ang_data.columns]
ang_cols=natsorted(ang_cols)
new_ang_data=new_ang_data[ang_cols]
new_ang_data=pd.concat([data[meta_cols],new_ang_data],axis=1)
print(new_ang_data.head())
```
